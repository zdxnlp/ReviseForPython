Python 的数据类型是编程的基础，它定义了数据的存储方式和可执行的操作。Python 中的数据类型主要分为基本数据类型（内置简单类型）和复合数据类型（容器类型）两大类，下面我会逐一详细讲解并给出实用示例。
# 基本数据类型

```python
# 正整数
age = 25
# 负整数
temperature = -5
# 零
zero = 0
# 超大整数（Python 支持）
big_num = 123456789012345678901234567890

# 类型验证
print(type(age))  # 输出：<class 'int'>
print(big_num + 1)  # 输出：123456789012345678901234567891
```

# 复合数据类型

1. 列表list
    - 有序、可变（可增删改元素）的序列，用[]包裹，元素可以是不同数据类型

```python
# 普通列表（元素类型混合）
fruits = ['apple', 'banana', 10, True]
# 空列表
empty_list = []
# 嵌套列表
matrix = [[1, 2], [3, 4]]

# 类型验证
print(type(fruits))  # 输出：<class 'list'>
# 修改元素（可变特性）
fruits[0] = 'orange'
print(fruits)  # 输出：['orange', 'banana', 10, True]
# 添加元素
fruits.append('grape')
print(fruits)  # 输出：['orange', 'banana', 10, True, 'grape']
```

2. 元组tuple
    - 有序、不可变（一旦创建无法修改元素）的序列，用()包裹

```python
# 普通元组
point = (10, 20)
# 省略括号的元组
colors = 'red', 'green', 'blue'
# 单元素元组（必须加逗号，否则视为普通值）
single_tuple = (5,)
# 空元组
empty_tuple = ()

# 类型验证
print(type(point))  # 输出：<class 'tuple'>
# 不可修改（修改会报错）
# point[0] = 15  # 报错：TypeError: 'tuple' object does not support item assignment
# 访问元素
print(point[1])  # 输出：20
```

3. 字典dict
    - 无序（Python3.7+后有序）、可变的键值对集合，用{}包裹，键(key)必须是不可变类型(如str/int/tuple)，值(value)可以是任意类型

```python
# 普通字典
user = {
    'name': '李四',
    'age': 30,
    'is_student': False
}
# 空字典
empty_dict = {}
# 嵌套字典
class_info = {
    'class1': {'num': 40, 'teacher': '王老师'},
    'class2': {'num': 38, 'teacher': '李老师'}
}

# 类型验证
print(type(user))  # 输出：<class 'dict'>
# 修改值（可变特性）
user['age'] = 31
print(user)  # 输出：{'name': '李四', 'age': 31, 'is_student': False}
# 获取值
print(user['name'])  # 输出：李四

print(class_info['class1'])
print(class_info['class1']['teacher'])
```

4. 集合set
    - 无序、不重复的元素集合，用{}包裹（空集合必须用set(),而非{}），元素必须是不可变类型。

```python
# 普通集合（自动去重）
nums = {1, 2, 2, 3, 3, 3}
print(nums)  # 输出：{1, 2, 3}
# 空集合
empty_set = set()
# 集合运算
a = {1, 2, 3}
b = {2, 3, 4}
print(a & b)  # 交集：{2, 3}
print(a | b)  # 并集：{1, 2, 3, 4}
print(a - b)  # 差集：{1}

# 类型验证
print(type(nums))  # 输出：<class 'set'>
```

# 不可变数据类型
- 定义：一旦创建，**其内部的值无法被直接修改**。如果想改变这个变量的值，本质是创建一个新的对象，然后让变量指向这个新对象（原对象如果没有其他引用，会被python垃圾回收）
- int、float、bool（本质是特殊的int）、str、tuple、NoneType
# 可变数据类型
- 定义：创建后，**可以直接修改其内部的值**，而不需要创建新对象（变量的内存地址不变）。
- list、dict、set

```python
# 定义字符串（不可变）
s = "hello"
# 打印初始内存地址（id() 函数返回对象的唯一标识）
print("初始地址：", id(s))

# 尝试"修改"字符串（实际是创建新对象）
s = s + " world"
# 打印修改后的内存地址
print("修改后地址：", id(s))

# 再以 int 为例
num = 10
print("num 初始地址：", id(num))
num = num + 5
print("num 修改后地址：", id(num))
```

```python
# 定义列表（可变）
lst = [1, 2, 3]
# 打印初始内存地址
print("列表初始地址：", id(lst))

# 直接修改列表内部元素（内存地址不变）
lst[0] = 100
print("列表修改后地址：", id(lst))

# 添加元素（同样不改变地址）
lst.append(4)
print("添加元素后地址：", id(lst))
```

- 元组的“看似可变”陷阱：元组本身不可变，但如果元组里包含可变类型元素，这个元素的内部值可以修改（元组的内存地址仍不变）

```python
t = (1, [2, 3])
print(id(t))
print(id(t[0]))
print(id(t[1]))
# t[0] = 100  # 报错（元组本身不可变）
t[1][0] = 200  # 可行（列表元素可变）
print(t)  # 输出：(1, [200, 3])
print(id(t))
print(id(t[0]))
print(id(t[1]))
```

- 函数参数传递的影响：
1. 不可变类型作为参数：函数内修改参数，不会影响外部变量（因为是新对象）
2. 可变类型作为参数：函数内部修改参数，会影响外部变量（因为是同一个对象）

```python
# 定义函数：尝试修改整数参数
def modify_immutable(num):
    print(f"函数内修改前，num 的地址：{id(num)}")  # 打印参数地址
    num = num + 10  # 看似修改，实际创建新对象
    print(f"函数内修改后，num 的地址：{id(num)}")  # 地址已变
    print(f"函数内 num 的值：{num}")

# 外部定义变量
external_num = 5
print(f"函数调用前，external_num 的地址：{id(external_num)}")
print(f"函数调用前，external_num 的值：{external_num}")

# 调用函数
modify_immutable(external_num)

# 查看外部变量是否变化
print(f"函数调用后，external_num 的值：{external_num}")
```

```python
# 定义函数：尝试修改列表参数
def modify_mutable(lst):
    print(f"函数内修改前，lst 的地址：{id(lst)}")  # 打印参数地址
    lst.append(4)  # 直接修改列表内部（不创建新对象）
    lst[0] = 100   # 直接修改列表元素
    print(f"函数内修改后，lst 的地址：{id(lst)}")  # 地址不变
    print(f"函数内 lst 的值：{lst}")

# 外部定义列表
external_list = [1, 2, 3]
print(f"函数调用前，external_list 的地址：{id(external_list)}")
print(f"函数调用前，external_list 的值：{external_list}")

# 调用函数
modify_mutable(external_list)

# 查看外部列表是否变化
print(f"函数调用后，external_list 的值：{external_list}")
```
