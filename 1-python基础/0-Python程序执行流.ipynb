```python
# 自定义元类，控制类的创建流程
class MyMeta(type):
    @classmethod
    def __prepare__(cls, name, bases, **kwargs):
        print(f"【类创建-1】调用元类__prepare__，类名：{name}")
        return super().__prepare__(name, bases, **kwargs)

    def __new__(cls, name, bases, namespace, **kwargs):
        print(f"【类创建-2】调用元类__new__，开始创建类对象：{name}")
        class_obj = super().__new__(cls, name, bases, namespace, **kwargs)
        print(f"【类创建-3】调用元类__new__，类对象创建完成：{name}")
        return class_obj

    def __init__(self, name, bases, namespace, **kwargs):
        print(f"【类创建-5】调用元类__init__，初始化类对象：{name}")
        super().__init__(name, bases, namespace, **kwargs)

    def __call__(self, *args, **kwargs):
        print(f"\n【实例创建-1】调用元类__call__，开始实例化{self.__name__}")
        # 调用类的__new__创建实例
        instance = self.__new__(self, *args, **kwargs)
        # 仅当实例是当前类的实例时，调用__init__
        if isinstance(instance, self):
            self.__init__(instance, *args, **kwargs)
        print(f"【实例创建-4】调用元类__call__，实例化完成")
        return instance

# 父类，用于演示__init_subclass__
class Base:
    def __init_subclass__(cls, **kwargs):
        print(f"【类创建-4】调用父类Base的__init_subclass__，子类：{cls.__name__}")
        super().__init_subclass__(**kwargs)

# 定义类，触发【类的初始化流程】
print("===== 开始定义类 =====")
class MyClass(Base, metaclass=MyMeta):
    print("【类创建-执行类体代码】开始执行类体")
    class_attr = 100
    def __new__(cls, *args, **kwargs):
        print(f"【实例创建-2】调用类的__new__，创建空实例")
        return super().__new__(cls)

    def __init__(self, value):
        print(f"【实例创建-3】调用类的__init__，初始化实例属性")
        self.value = value
    print("【类创建-执行类体代码】类体执行完成")
print("===== 类定义完成 =====\n")

# 实例化，触发【对象的初始化流程】
print("===== 开始实例化 =====")
obj = MyClass(200)
print(f"===== 实例化完成，实例属性：{obj.value} =====")
```
