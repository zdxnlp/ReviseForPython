# 1.列表
- 有序、可变、可重复的序列数据结构
1. 定义：用[]包裹，元素间用逗号分隔；也可以通过list()函数创建；元素可以是不同数据类型（数字、字符串、列表等），支持增删改查。

```python
# 基本定义
empty_list = []  # 空列表
num_list = [1, 2, 3, 4]  # 同类型元素
mix_list = [1, "hello", True, [5, 6]]  # 不同类型元素（包含嵌套列表）

# 通过list()创建
str_to_list = list("python")  # 字符串转列表：['p', 'y', 't', 'h', 'o', 'n']
tuple_to_list = list((7, 8, 9))  # 元组转列表：[7, 8, 9]

print(empty_list)
print(num_list)
print(mix_list)
print(str_to_list)
```

2. 切片：列表是有序序列，支持切片（截取部分元素），返回新列表，不修改原列表
- 列表[start:end:step]
- start:起始索引（包含，默认0）
- end:结束索引（不包含，默认列表长度）
- step:步长（默认1，负数表示反向截取）
- 索引规则：正向索引从0开始，反向索引从-1开始（最后一个元素为-1）

```python
lst = [10, 20, 30, 40, 50]

# 基础切片
print(lst[1:4])    # [20, 30, 40] （取索引1到3）
print(lst[:3])     # [10, 20, 30] （省略start，从0开始）
print(lst[2:])     # [30, 40, 50] （省略end，到末尾）
print(lst[:])      # [10, 20, 30, 40, 50] （复制整个列表）

# 带步长切片
print(lst[0:5:2])  # [10, 30, 50] （每隔1个取1个）
print(lst[::-1])   # [50, 40, 30, 20, 10] （反向截取，常用作列表反转）

# 反向索引切片
print(lst[-3:-1])  # [30, 40] （取倒数第3到倒数第2个）
```

3. 常用接口
- append(x):末尾添加元素x
- extend(x):末尾批量添加可迭代对象（列表、元组等）
- insert(i,x):在索引i处插入元素x
- remove(x):删除第一个匹配的元素x（无则报错）
- pop(i):删除索引i处元素（默认最后一个），返回该元素
- del list\[i]:直接删除索引i处元素（也可切片删除）
- clear(():清空列表
- index(x):返回第一个匹配x的索引（无则报错）
- count(x):统计元素x出现的次数
- sort(x):原地排序（默认升序，reverse=True降序）
- reverse():原地反转列表
- len(lst):获取列表长度
- x in lst:判断元素x是否在列表中（返回bool值）
- lst\[i] = new_val:直接修改索引i处的元素

```python
lst = [10, 20, 30, 40, 50]
print(len(lst))
print(20 in lst)
lst.append(60)
print(lst)
lst.extend([70,80])
print(lst)
lst.remove(30)
print(lst)
lst.insert(2, 25)
print(lst)
res = lst.pop(-1)
print(res)
del lst[0:2]
print(lst)
print(lst.index(40))
print([1, 2, 2, 3].count(2))
lst.sort(reverse=True)
print(lst)
lst.reverse()
print(lst)
lst.clear()
print(lst)
```

# 2.元组
元组是有序、不可变、可重复的序列数据结构，适合存储无序修改的固定数据，性能比列表高。
1. 定义
- 语法：用()包裹元素，元素间用逗号分隔；也可通过tuple()函数创建。
- 空元组用()；单元素元组必须加逗号(5,)，否则会被识别为普通变量。

```python
# 基本定义
empty_tuple = ()  # 空元组
num_tuple = (1, 2, 3, 4)  # 同类型元素
mix_tuple = (1, "hello", True, (5, 6))  # 嵌套元组
print(mix_tuple)

# 特殊定义（省略括号）
simple_tuple = 10, 20, 30  # 等价于 (10,20,30)
single_tuple = (5,)  # 单元素元组（必须加逗号）
print(type(single_tuple))
wrong_tuple = (5)    # 不是元组，是整数 5
print(type(wrong_tuple))

# 通过tuple()创建
list_to_tuple = tuple([1,2,3])  # 列表转元组：(1,2,3)
str_to_tuple = tuple("python")  # 字符串转元组：('p','y','t','h','o','n')
print(list_to_tuple)
print(str_to_tuple)
```

2. 切片：元组和列表一样支持切片（因为有序），但切片返回的是新元组（不可变，无法修改切片结果）

```python
tup = (10, 20, 30, 40, 50)
print(tup[1:4])    # (20, 30, 40)
print(tup[::-1])   # (50, 40, 30, 20, 10) （反向切片）
print(tup[:3])     # (10, 20, 30)
```

3. 常用接口：元组不可变，因此没有增删改的方法，仅提供查询类接口
- index(x):返回第一个匹配x的索引，无则报错
- count(x):统计x元素出现的次数
- len(tup):获取元组的长度
- x in tup:判断元素是否存在
- 元组拆包：a,b,c = (10,20,30)->a=10,b=20,c=30（常用作函数多返回值）

```python
print(tup.index(40))
print((1,2,2,3).count(2))
```

# 3.集合
集合是无序、可变（但元素不可变）、无重复的无序数据结构，适合去重、集合运算（交集、并集）等。
1. 定义
- 语法：用{}包裹元素（注意空集合不能用{}，需用set()）;也可以通过set()函数创建。
- 限制：元素必须是不可变类型（数字、字符串、元组），不能是列表、集合等可变类型（否则报错）。

```python
# 基本定义
empty_set = set()  # 空集合（必须用set()，{}是空字典）
print(type(empty_set))
num_set = {1, 2, 3, 4}  # 同类型元素
print(type(num_set))
print(num_set)
mix_set = {1, "hello", (5, 6)}  # 元素为不可变类型
print(type(mix_set))
print(mix_set)

# 自动去重特性
dup_set = {1, 2, 2, 3, 3}  # 自动去重 → {1,2,3}
print(dup_set)

# 通过set()创建
list_to_set = set([1,2,2,3])  # 列表去重 → {1,2,3}
print(list_to_set)
str_to_set = set("python")    # 字符串转集合 → {'p','y','t','h','o','n'}（无序）
print(str_to_set)
```

2. 切片：集合是无序的，没有索引概念，因此不支持切片操作

3. 常用接口
- add(x):添加元素x ，x必须不可变，重复则忽略
- update(iter):批量添加可迭代对象（自动去重）
- remove(x):删除元素x，无则报错
- discard(x):删除元素x，无则忽略
- pop(x):随机删除一个元素，并返回该元素
- clear():清空集合
- issubset(s2):判断是否是s2的子集
- issuperset(s2):判断是否包含s2（超集）
- len(s):获取集合长度
- x in s:判断元素是否存在
- 集合不支持索引访问：s\[0]报错

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

# 交集（共同元素）：& 或 intersection()
print(s1 & s2)          # {3,4}
print(s1.intersection(s2))  # 同上

# 并集（所有元素，去重）：| 或 union()
print(s1 | s2)          # {1,2,3,4,5,6}

# 差集（s1有、s2无）：- 或 difference()
print(s1 - s2)          # {1,2}

# 对称差集（仅一方有）：^ 或 symmetric_difference()
print(s1 ^ s2)          # {1,2,5,6}
```

# 4.字典
字典是python中核心的键值对（key-value）数据结构，3.7+版本后默认有序（基于插入顺序）、可变，键（key）唯一且不可变，值（value）可任意，适合通过“键”快速查找“值”（查询效率远高于列表）
1. 定义
- 语法：用{}包裹键值对（key:value形式），键值对间用逗号分隔；也可通过dict()函数创建。
- 核心规则：
    - 键必须是不可变类型（数字、字符串、元组），不能是列表、集合等可变类型，否则报错；
    - 键唯一，重复定义会覆盖原有值；
    - 值无类型限制，可是任意数据（数字、字符串、列表、字典等）。

```python
# 基本定义
empty_dict = {}  # 空字典
print(type(empty_dict))
basic_dict = {"name": "Alice", "age": 20, "city": "Beijing"}  # 常规字典
print(basic_dict)
mix_dict = {
    "id": 101,
    "scores": [85, 90, 95],  # 值为列表
    "info": ("female", 1.65),  # 值为元组
    "hobby": {"reading", "running"}  # 值为集合
}
print(mix_dict)

# 特殊键类型（不可变）
valid_key_dict = {
    123: "数字键",
    "name": "字符串键",
    (1, 2): "元组键"  # 元组作为键（合法）
}
print(valid_key_dict)
# invalid_key_dict = {[1,2]: "列表键"}  # 报错：TypeError（列表是可变类型，不能作为键）

# 通过dict()创建
dict_from_tuple = dict([("name", "Bob"), ("age", 22)])  # 列表嵌套元组 → {'name':'Bob','age':22}
print(dict_from_tuple)
dict_from_kwargs = dict(name="Charlie", age=25)  # 关键字参数 → {'name':'Charlie','age':25}
print(dict_from_kwargs)

# 键重复会覆盖
dup_key_dict = {"name": "Dave", "name": "Eve"}  # 最终 → {'name': 'Eve'}
print(dup_key_dict)
```

2. 切片：不支持。字典是键值对映射结构，无“索引”概念（即使3.7+有序，也不能通过索引访问），因此不支持切片操作。
3. 常用接口：字典的核心操作围绕“键值对”的增删改查
- dict\[key] = value:添加 / 修改键值对（key 存在则修改，不存在则添加）
- dict\[key]:通过键访问值（key 不存在则报错
- dict.get(key, default):通过键访问值（key 不存在返回 default，默认 None，推荐使用）
- dict.pop(key, default):删除指定键的键值对，返回对应值（key 不存在则返回 default，否则报错）
- del dict\[key]:直接删除指定键的键值对（key 不存在则报错）
- dict.update(new_dict):批量添加 / 修改键值对（合并字典，重复键覆盖）
- dict.keys():返回所有键的可迭代对象（视图），可转列表
- dict.values():返回所有值的可迭代对象（视图），可转列表
- dict.items():返回所有键值对的可迭代对象（(key,value) 元组），可转列表
- dict.clear():清空字典
- dict.copy():浅拷贝字典（生成新字典，避免修改原字典）
- dict.setdefault(key, default):若 key 不存在则添加（值为 default），存在则返回原值
- len(dict)：获取字典中键值对的数量；
- key in dict：判断键是否存在（仅判断键，不判断值）；
- 遍历字典：for key in dict（遍历键）、for key, value in dict.items()（遍历键值对）。

```python
print(basic_dict)
basic_dict["age"] = 21
print(basic_dict)
basic_dict["name"]
```

```python
basic_dict.get("gender", "unknown")
```

```python
basic_dict.pop("city")
print(basic_dict)
```

```python
del basic_dict["age"]
print(basic_dict)
```

```python
basic_dict.update({"age":20, "gender":"female"})
print(basic_dict)
```

```python
list(basic_dict.keys())
```

```python
list(basic_dict.values())
```

```python
list(basic_dict.items())
```

```python
new_dict = basic_dict.copy()
print(new_dict)
```

```python
basic_dict.clear()
print(basic_dict)
```

```python
basic_dict.setdefault("gender", "female")
print(basic_dict)
```

```python
# 遍历示例
person = {"name": "Alice", "age": 20}
for k in person:
    print(k, person[k])  # 输出：name Alice → age 20

for k, v in person.items():
    print(f"{k}: {v}")  # 更优雅的遍历键值对
```

# 比较


- 列表：需要频繁添加 / 删除 / 修改元素，且需保持顺序（如用户列表、商品列表、日志记录）；
- 元组：数据固定不变（如坐标 (x,y)、函数多返回值、配置参数），或需要作为字典的键 / 集合的元素；
- 集合：需要去重（如列表去重）、判断元素是否存在（高效）、集合运算（如找两个列表的共同元素）；
- 字典：需要通过 “唯一标识”（键）快速查找数据（如用户信息、配置项、缓存数据）。

- 元组的 “不可变” 是指元素本身不可修改，但如果元素是可变类型（如列表），则该元素内部可修改：

```python
tup = (1, [2, 3])
print(tup)
tup[1].append(4)  # 合法，元组变为 (1, [2,3,4])
print(tup)
```

- 集合和字典的 “高效查询” 基于哈希表，因此键 / 元素必须是可哈希的（不可变类型）；
- 3.7 之前的字典是无序的，若需兼容旧版本，不要依赖字典的顺序特性。
