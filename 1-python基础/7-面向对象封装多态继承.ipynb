# 封装
1. 定义：封装是将对象的属性（数据） 和方法（操作数据的函数） 绑定在一起，隐藏内部实现细节，只对外暴露统一、安全的访问接口。
- 在 Python 中，通过给属性 / 方法加 __（双下划线）标记为私有成员，外部无法直接访问；通过定义公共方法（如 get_xxx/set_xxx）作为访问入口。
- 核心作用：
    - 数据安全：防止外部随意修改核心属性，避免数据被破坏（比如禁止直接修改银行账户余额）；
    - 降低耦合：外部只需关注接口功能，无需了解内部实现，便于维护和修改；
    - 代码复用：属性和方法封装在类中，实例化后可直接复用。

```python
class BankAccount:
    def __init__(self, account_name):
        self.account_name = account_name  # 公开属性（允许外部查看）
        self.__balance = 0  # 私有属性（双下划线开头，外部无法直接访问）

    # 公共方法：存款（带合法性校验）
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"{self.account_name}存款{amount}元，当前余额：{self.__balance}元")
        else:
            print("存款金额必须大于0！")

    # 公共方法：取款（带余额校验）
    def withdraw(self, amount):
        if amount <= 0:
            print("取款金额必须大于0！")
        elif amount > self.__balance:
            print(f"余额不足！当前余额：{self.__balance}元")
        else:
            self.__balance -= amount
            print(f"{self.account_name}取款{amount}元，当前余额：{self.__balance}元")

    # 公共方法：查询余额（只读接口）
    def get_balance(self):
        return self.__balance

# 测试封装效果
if __name__ == "__main__":
    account = BankAccount("张三")
    account.deposit(1000)  # 张三存款1000元，当前余额：1000元
    account.withdraw(500)  # 张三取款500元，当前余额：500元

    # 尝试直接访问私有属性（会报错）
    # print(account.__balance)  # AttributeError: 'BankAccount' object has no attribute '__balance'

    # 只能通过公共方法查询余额
    print(f"当前余额：{account.get_balance()}元")  # 当前余额：500元

    # 非法操作被拦截
    account.withdraw(600)  # 余额不足！当前余额：500元
```

- 关键解释：__balance 是私有属性，外部无法直接修改，必须通过 deposit/withdraw 方法操作，这两个方法会校验金额合法性，保证余额不会被随意篡改，体现了封装的安全性。

# 继承
1. 定义：继承允许一个类（子类 / 派生类）继承另一个类（父类 / 基类）的属性和方法，子类可以直接复用父类的功能，也可以重写 / 扩展父类的方法。
    - Python 支持单继承（一个子类继承一个父类）和多继承（一个子类继承多个父类）。
2. 核心作用：
- 代码复用：避免重复编写相同的属性 / 方法，子类直接 “继承” 父类的功能；
- 层次化设计：建立类之间的 “is-a” 关系（如 Dog is a Animal），让代码结构更清晰；
- 灵活扩展：子类可在父类基础上新增功能，无需修改父类代码。
3. 关键语法：
- 定义子类：class 子类名(父类名):
- 调用父类方法：super().父类方法()（推荐，自动适配继承关系）
- 多继承：class 子类名(父类1, 父类2):
## 示例1：单继承

```python
# 父类：动物（通用属性和方法）
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name}正在吃东西...")

# 子类：狗（继承Animal，扩展独有方法）
class Dog(Animal):
    # 扩展子类独有的方法
    def bark(self):
        print(f"{self.name}汪汪叫！")

    # 重写父类方法（定制化）
    def eat(self):
        print(f"{self.name}正在啃骨头...")

# 子类：猫（继承Animal，扩展独有方法）
class Cat(Animal):
    def meow(self):
        print(f"{self.name}喵喵叫！")

# 测试
if __name__ == "__main__":
    dog = Dog("旺财")
    dog.eat()   # 旺财正在啃骨头...（重写后的方法）
    dog.bark()  # 旺财汪汪叫！（子类扩展方法）

    cat = Cat("咪咪")
    cat.eat()   # 咪咪正在吃东西...（继承父类方法）
    cat.meow()  # 咪咪喵喵叫！（子类扩展方法）
```

## 示例2:带super()的继承——初始化父类属性

```python
# 父类：人
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"我叫{self.name}，今年{self.age}岁。")

# 子类：学生（继承Person，扩展学号属性）
class Student(Person):
    def __init__(self, name, age, student_id):
        # 调用父类的__init__，初始化name和age
        super().__init__(name, age)
        # 子类扩展属性
        self.student_id = student_id

    # 重写父类方法，扩展功能
    def introduce(self):
        super().introduce()  # 先执行父类的introduce
        print(f"我的学号是{self.student_id}。")

# 测试
student = Student("李四", 18, "2024001")
student.introduce()
# 输出：
# 我叫李四，今年18岁。
# 我的学号是2024001。
```

## 示例3:多继承——同时继承多个父类

```python
# 父类1：会跑
class Runnable:
    def run(self):
        print("正在奔跑...")

# 父类2：会飞
class Flyable:
    def fly(self):
        print("正在飞翔...")

# 子类：鸟（继承两个父类）
class Bird(Runnable, Flyable):
    def __init__(self, name):
        self.name = name

# 测试
bird = Bird("麻雀")
bird.run()  # 正在奔跑...（继承Runnable）
bird.fly()  # 正在飞翔...（继承Flyable）
```

# 多态
1. 定义：多态指 “同一个接口，不同实现”：不同类的对象调用同一个方法名，会根据对象的类型自动执行对应的逻辑。
    - 多态的实现依赖于方法重写（子类重写父类的方法）。
2. 核心作用：
    - 提高灵活性：无需关注对象具体类型，只需调用统一接口；
    - 符合开闭原则：新增类时，无需修改原有调用逻辑，只需实现统一接口；
    - 简化逻辑：统一的调用方式，降低代码复杂度。
## 示例1:基础多态——不同动物的叫声

```python
# 父类：定义统一接口
class Animal:
    def __init__(self, name):
        self.name = name

    # 统一接口：发出声音（父类占位，子类重写）
    def make_sound(self):
        pass

# 子类：狗
class Dog(Animal):
    def make_sound(self):
        return f"{self.name}：汪汪汪！"

# 子类：猫
class Cat(Animal):
    def make_sound(self):
        return f"{self.name}：喵喵喵！"

# 子类：鸭子
class Duck(Animal):
    def make_sound(self):
        return f"{self.name}：嘎嘎嘎！"

# 统一调用函数（无需区分动物类型）
def animal_sound(animal):
    print(animal.make_sound())

# 测试
if __name__ == "__main__":
    dog = Dog("旺财")
    cat = Cat("咪咪")
    duck = Duck("唐老鸭")

    # 调用同一函数，自动执行对应类的方法
    animal_sound(dog)   # 旺财：汪汪汪！
    animal_sound(cat)   # 咪咪：喵喵喵！
    animal_sound(duck)  # 唐老鸭：嘎嘎嘎！

    # 新增动物类，无需修改animal_sound（开闭原则）
    class Cow(Animal):
        def make_sound(self):
            return f"{self.name}：哞哞哞！"

    animal_sound(Cow("小花"))  # 小花：哞哞哞！
```

## 示例2:多态的实际应用

```python
# 父类：支付方式（统一接口）
class Payment:
    def pay(self, amount):
        """统一支付接口，子类必须实现"""
        raise NotImplementedError("子类必须重写pay方法")

# 子类：微信支付
class WeChatPay(Payment):
    def pay(self, amount):
        print(f"使用微信支付{amount}元，支付成功！")

# 子类：支付宝支付
class AliPay(Payment):
    def pay(self, amount):
        print(f"使用支付宝支付{amount}元，支付成功！")

# 统一支付处理函数
def process_payment(payment_method, amount):
    if amount <= 0:
        print("支付金额必须大于0！")
        return
    payment_method.pay(amount)

# 测试
wechat = WeChatPay()
process_payment(wechat, 100)  # 使用微信支付100元，支付成功！

ali = AliPay()
process_payment(ali, 200)     # 使用支付宝支付200元，支付成功！

# 新增银行卡支付，无需修改process_payment
class BankPay(Payment):
    def pay(self, amount):
        print(f"使用银行卡支付{amount}元，支付成功！")

process_payment(BankPay(), 300)  # 使用银行卡支付300元，支付成功！
```
