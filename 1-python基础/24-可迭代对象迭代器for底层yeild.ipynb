# 1、什么是可迭代对象？
- 可以被for循环逐个取出元素的对象
- list,tuple,str,dict,set,range()

````python
nums = [1, 2, 3]
# nums就是可迭代对象
for n in nums:
    print(n)
````

````python
for ch in "abc":
    print(ch)
````

````python
for key in {"name": "Tom", "age": 18}:
    print(key)
````

# 2、iter()是干什么的？
- 可迭代对象本身只是“能被遍历”，但真正负责“一个一个取值”的，是**迭代器**。
- iter()的作用就是：把可迭代对象，变成迭代器。

````python
nums = [10, 20, 30] # nums是可迭代对象
it = iter(nums) # it是迭代器
````

- 如果一个对象能传给iter()，通常它就是可迭代对象。

````python
iter([1, 2, 3])   # 可以
iter("hello")     # 可以
iter(100)         # 不可以 TypeError: 'int' object is not iterable
````

# 3、什么是迭代器
- 迭代器是一个能“记住当前位置，并且每次给出下一个值”的对象。
- 迭代器最大的特点是可以配合next()使用。

````python
nums = [10, 20, 30]
it = iter(nums)

print(next(it))  # 10
print(next(it))  # 20
print(next(it))  # 30
````

- 如果继续取

````python
print(next(it)) # StopIteration:
````

- one word：迭代器=可以被next()一个一个取值的对象。

# 4、可迭代对象和迭代器的区别
- 可迭代对象：能拿来遍历
- 迭代器：能一个一个往后取，并记住取到哪里

````python
nums = [1,2,3]
it = iter(nums)
# nums是可迭代对象
# it才是迭代器
````

- 可迭代对象->iter()->迭代器

# 5、整条链：可迭代对象 -> iter() -> 迭代器 -> next()
- 这是python迭代机制最核心的一条链。

1. 可迭代对象：里面有很多元素
2. iter()：拿到一个“负责取元素”的对象
3. 迭代器：负责按顺序记录和取值
4. next()：每次向前取一个值

# 6、for循环底层做了什么？

````python
for x in [1, 2, 3]:
    print(x)
````

- 底层逻辑大致相当于：

````python
it = iter([1, 2, 3])

while True:
    try:
        x = next(it)
        print(x)
    except StopIteration:
        break
````

- 也就是说，for循环会自动帮我们做这几件事：
1. 嗲用iter()拿到迭代器
2. 反复调用next()
3. 每次拿到一个值就执行循环体
4. 遇到StopIteration时结束循环
- **for item in obj**的本质，就是不断对 iter(obj) 调用 next()

# 7、什么是生成器？
- 生成器是一种特殊的对象，它既是可迭代对象，也是迭代器。

````python
gen = (x * 2 for x in range(3))
````

- 这个gen就是生成器，它可以被for遍历：

````python
for x in gen:
    print(x)
````

- 也可以用next()一个一个取:

````python
gen = (x * 2 for x in range(3))

print(next(gen))  # 0
print(next(gen))  # 2
print(next(gen))  # 4
````

- 所以：
1. 列表是可迭代对象，但不是迭代器
2. 生成器既是可迭代对象，也是迭代器

# 8、什么是yeild？
- yeild是用来“暂停函数并返回一个值”的关键字
- 它和return的区别是：
1. return：返回一次后，函数彻底结束
2. yeild：返回一个值后，函数暂停，下次还能从暂停处继续执行

````python
def foo():
    yield 1
    yield 2
    yield 3

f = foo()
````

- f是一个生成器

````python
print(next(f))  # 1
print(next(f))  # 2
print(next(f))  # 3
````

- 只要函数里出现了yeild，它就变成了生成器函数。

# 9、yeild为什么重要？
- 它能让函数“一个一个地产生数据”，而不是一次性把所有结果都放进内存

````python
def my_range(n):
    i = 0
    while i < n:
        yield i
        i += 1
````

````python
for x in my_range(5):
    print(x)
````

- 这类写法适合处理大量数据，因为他是“边要边生产”
- yield = 返回一个值，但保留现场，下次继续。

# 10、自己怎么写一个可迭代对象
- 最简单的方法是给类写\_\_iter__()

````python
class MyNumbers:
    def __init__(self, n):
        self.n = n

    def __iter__(self):
        for i in range(self.n):
            yield i
````

````python
obj = MyNumbers(5)

for x in obj:
    print(x)
````

- 这里 obj 就是你自己写的可迭代对象。因为 for 循环会调用它的 \_\_iter__()，而 yield 生成了迭代器。

# 11、怎么写一个迭代器？
- 自己实现 \_\_next__()

````python
class MyIterator:
    def __init__(self, n):
        self.n = n
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.n:
            value = self.current
            self.current += 1
            return value
        else:
            raise StopIteration
````

````python
obj = MyIterator(5)

for x in obj:
    print(x)
````
