eval() 是 Python 的内置函数，核心作用是执行一个字符串形式的 Python 表达式，并返回表达式的执行结果。

# 一、基本语法
eval(expression, globals=None, locals=None)
- expression：必填，字符串形式的 Python 表达式（注意是 “表达式”，不是完整语句如赋值、循环）。
- globals：可选，字典类型，指定表达式执行时的全局命名空间（即可用的全局变量 / 函数）。
- locals：可选，字典类型，指定表达式执行时的局部命名空间。

# 二、基础示例
1. 执行数学表达式

```python
result = eval("1 + 2 * 3")
print(result)  # 输出：7
```

2. 使用变量

```python
x = 5
y = 10
result = eval("x + y * 2")
print(result)  # 输出：25
```

3. 处理字面量（列表、字典等）

```python
# 解析字符串形式的列表
list_str = "[1, 2, 3, 4]"
my_list = eval(list_str)
print(my_list)  # 输出：[1, 2, 3, 4]

# 解析字符串形式的字典
dict_str = '{"name": "Alice", "age": 20}'
my_dict = eval(dict_str)
print(my_dict)  # 输出：{'name': 'Alice', 'age': 20}
```

4. 配合 globals/locals 限制命名空间

```python
# 仅允许使用指定的全局变量
result = eval("a + b", globals={"a": 10, "b": 20})
print(result)  # 输出：30

# 结合全局和局部命名空间
result = eval("x + y", globals={"x": 5}, locals={"y": 3})
print(result)  # 输出：8
```

# 三、核心限制：仅支持“表达式”
eval() 只能执行表达式（会返回值的代码），不能执行语句（如赋值、循环、try-except 等）：

```python
# 错误示例：赋值是语句，不是表达式
eval("a = 1")  # 报错：SyntaxError: invalid syntax

# 错误示例：循环是语句
eval("for i in range(5): print(i)")  # 报错
```

# 四、安全隐患
eval() 会无条件执行字符串中的任意代码，如果字符串来自不可信的输入（如用户输入、网络数据），会造成严重的安全问题！
- 恶意代码示例：

```python
# 示例 1：执行系统命令（Windows 下列出目录，Linux 下同理）
eval("__import__('os').system('dir')")

# 示例 2：删除文件（危险！仅演示，请勿执行）
# eval("__import__('os').remove('important_file.txt')")
```

# 五、安全替代方案：ast.literal_eval()
如果仅需解析字面量（字符串、数字、列表、字典、元组、布尔值、None），推荐使用 ast.literal_eval()，它不会执行任意代码，更安全：

```python
import ast

# 安全解析列表
list_str = "[1, 2, 3]"
my_list = ast.literal_eval(list_str)
print(my_list)  # 输出：[1, 2, 3]

# 安全解析字典
dict_str = '{"name": "Bob"}'
my_dict = ast.literal_eval(dict_str)
print(my_dict)  # 输出：{'name': 'Bob'}

# 尝试执行表达式会报错（安全）
# ast.literal_eval("1 + 2")  # 报错：ValueError: malformed node or string
```
