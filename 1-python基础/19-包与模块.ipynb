# Python 包与模块详解
- 在 Python 中，模块（Module） 和 包（Package） 是组织代码的核心方式，用于实现代码复用、避免命名冲突，并让项目结构更清晰。
# 一、模块（Module）
1. 什么是模块？
    - 一个 .py 文件就是一个模块。模块中可以定义函数、类、变量，也可以包含可执行代码。
2. 模块的作用
    - 代码复用：将常用功能封装在模块中，其他文件可直接导入使用，无需重复编写。
    - 逻辑组织：将相关功能放在同一个模块中，便于维护和理解。
3. 创建于使用模块

- 步骤1:创建模块 math_utils.py

```python
# math_utils.py（这是一个模块）
PI = 3.14159  # 模块级变量

def add(a, b):
    return a + b

def circle_area(radius):
    return PI * radius ** 2
```

- 步骤 2：在另一个文件中导入模块

```python
# main.py
import math_utils  # 导入整个模块

print(math_utils.PI)  # 访问模块变量：3.14159
print(math_utils.add(1, 2))  # 调用模块函数：3
print(math_utils.circle_area(5))  # 78.53975

# 也可以只导入模块中的特定内容
from math_utils import add, PI
print(add(3, 4))  # 7
print(PI)  # 3.14159
```

# 二、包（Package）
1. 什么是包
    - 包是一个包含 \_\_init__.py 文件的目录，用于组织多个相关的模块，形成一个 “功能集合”。包可以嵌套（即包中包含子包）。
2. 包的作用
    - 避免命名冲突：不同包中的模块可以同名，通过 “包名。模块名” 区分。
    - 分层管理代码：适合大型项目，将功能按目录结构组织。
3. \_\_init__.py的作用：是包的标识文件（Python 3.3+ 虽支持 “命名空间包” 无需此文件，但普通包仍推荐保留），它可以：
    - 初始化包（如导入子模块、定义包级变量）。
    - 控制 from package import * 的行为（通过 \_\_all__ 列表指定导入内容）。
4. 创建与使用包

- 步骤1:构建包的目录结构

```python
"""
my_project/
├── main.py
└── my_package/          # 这是一个包
    ├── __init__.py      # 包的标识文件
    ├── math_utils.py    # 包中的模块
    └── string_utils.py  # 包中的另一个模块
"""
```

- 步骤2:编写包内模块

```python
# my_package/math_utils.py
def add(a, b):
    return a + b
```

```python
# my_package/string_utils.py
def to_upper(s):
    return s.upper()
```

- 步骤3:编写\_\_init__.py（可选，用于简化导入）

```python
# my_package/__init__.py
# 导入子模块，让外部可以直接用 my_package.add 而无需 my_package.math_utils.add
from .math_utils import add
from .string_utils import to_upper

# 控制 from my_package import * 的行为
__all__ = ['add', 'to_upper']
```

- 步骤4:在main.py中使用包

```python
# main.py
import my_package  # 导入整个包

print(my_package.add(1, 2))  # 3（通过 __init__.py 简化了导入）
print(my_package.to_upper("hello"))  # HELLO

# 也可以直接导入包中的模块
from my_package import math_utils
print(math_utils.add(3, 4))  # 7
```

# 三、程序运行时如何找到包与模块？
Python 导入模块时，会按照固定的搜索路径依次查找，找到第一个匹配的模块就停止。
1. 模块搜索顺序（优先级从高到低）
    1. 当前目录：脚本所在的目录（或交互式解释器的当前工作目录）。
    2. 环境变量 PYTHONPATH：用户预先设置的目录列表。
    3. Python 标准库目录：Python 安装时自带的库（如 os、sys）。
    4. 第三方库目录：site-packages（通过 pip 安装的库都在这里）。
2. 查看搜索路径：sys.path。Python 会将所有搜索路径存储在 sys.path 列表中，可以打印查看：

```python
import sys
print(sys.path)  # 输出一个目录列表，按搜索顺序排列
```

3. 动态修改搜索路径（临时生效）。如果模块不在默认搜索路径中，可以通过 sys.path.append() 临时添加：

```python
import sys
sys.path.append("/path/to/your/modules")  # 添加自定义目录到搜索路径末尾

import your_module  # 现在可以导入了
```

 注意：这种方式仅在当前程序运行时生效，关闭程序后失效。

- 模块：单个 .py 文件，用于封装可复用代码。
- 包：包含 \_\_init__.py 的目录，用于组织多个模块，避免命名冲突。
