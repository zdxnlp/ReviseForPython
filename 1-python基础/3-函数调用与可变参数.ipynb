# 参数传递

1. 位置参数

```python
def print_info(name, age):
    print(f"姓名：{name}，年龄：{age}")

# 位置参数：按顺序传参
print_info("张三", 25)  # 输出：姓名：张三，年龄：25
```

2. 关键字参数

```python
# 关键字参数：按参数名传参，顺序无关
print_info(age=25, name="张三")  # 输出：姓名：张三，年龄：25
```

3. 默认参数

```python
# 定义带默认参数的函数（默认参数放最后）
def print_info(name, age=18):
    print(f"姓名：{name}，年龄：{age}")

# 调用1：不传默认参数，使用默认值18
print_info("李四")  # 输出：姓名：李四，年龄：18
# 调用2：传默认参数，覆盖默认值
print_info("李四", 20)  # 输出：姓名：李四，年龄：20
```

# 返回值处理
- 无返回值：函数默认返回 None；
- 单个返回值：直接返回任意类型；
- 多个返回值：本质返回元组，可解构为多个变量。

```python
# 无返回值
def say_hello():
    print("Hello")

res1 = say_hello()
print(res1)  # 输出：None

# 单个返回值
def get_num():
    return 10

res2 = get_num()
print(res2)  # 输出：10

# 多个返回值（解构）
def get_info():
    return "张三", 25, "男"

name, age, gender = get_info()
print(name, age, gender)  # 输出：张三 25 男
```

# 调用函数时的作用域注意
- 函数调用时，优先使用函数内部的局部变量，若没有则查找全局变量：

```python
# 全局变量
num = 100

def func():
    # 局部变量（覆盖全局）
    num = 200
    print(num)  # 输出：200

func()
print(num)  # 输出：100（全局变量未被修改）
```

# 可变参数
可变参数允许函数接收任意数量的参数，分为两种：*args（可变位置参数）和 **kwargs（可变关键字参数），是 Python 函数灵活性的核心。
1. 可变位置参数（*args）
    - 定义：在参数名前加*，函数内部会将接收的任意数量位置参数打包成(元组)；
    - 用途：处理不确定数量的位置参数（求和、拼接多个值）

```python
def calc_sum(*args):
    """计算任意数量数字的和"""
    print(f"args 的类型：{type(args)}")  # 输出：<class 'tuple'>
    total = 0
    for num in args:
        total += num
    return total

# 调用1：传3个参数
print(calc_sum(1, 2, 3))  # 输出：6
# 调用2：传5个参数
print(calc_sum(10, 20, 30, 40, 50))  # 输出：150
# 调用3：不传参数（args 为空元组）
print(calc_sum())  # 输出：0
```

- 若已有列表/元组，可在前面加*解包后传给*args：

```python
nums = [1, 2, 3]
# 解包列表（等价于 calc_sum(1,2,3)）
print(calc_sum(*nums))  # 输出：6
```

2. 可变关键字参数（**kwargs）
    - 定义：在参数名前加**，函数内部会将接收的任意数量关键字参数打包成{字典}
    - 用途：处理不确定数量的关键字参数（接收任意配置项、用户信息）

```python
def print_user_info(**kwargs):
    """打印任意数量的用户信息"""
    print(f"kwargs 的类型：{type(kwargs)}")  # 输出：<class 'dict'>
    for key, value in kwargs.items():
        print(f"{key}：{value}")

# 调用1：传2个关键字参数
print_user_info(name="张三", age=25)
# 输出：
# name：张三
# age：25

# 调用2：传3个关键字参数
print_user_info(name="李四", age=30, gender="男")
# 输出：
# name：李四
# age：30
# gender：男
```

- 若已有字典，可在前面加 ** 解包后传给 **kwargs：

```python
user = {"name": "王五", "age": 28, "city": "北京"}
# 解包字典（等价于 print_user_info(name="王五", age=28, city="北京")）
print_user_info(**user)
```

3. 可变参数的使用规则
    - 若函数同时包含普通参数、默认参数、*args、**kwargs，需按以下顺序定义：普通参数 → 默认参数 → *args → **kwargs

```python
# 正确顺序
def func(a, b=10, *args, **kwargs):
    print(f"a={a}, b={b}, args={args}, kwargs={kwargs}")

# 调用
func(1, 2, 3, 4, x=5, y=6)
# 输出：a=1, b=2, args=(3, 4), kwargs={'x': 5, 'y': 6}
```

- 单独使用*分隔参数：在参数列表中单独写 *，表示后续参数必须用关键字参数传递：

```python
def func(a, *, b, c):
    print(a, b, c)

# 正确调用（b、c 必须用关键字）
func(1, b=2, c=3)  # 输出：1 2 3
# 错误调用（b 用位置参数）
# func(1, 2, 3)  # 报错：func() takes 1 positional argument but 3 were given
```

- 综合示例

```python
def create_report(title, type="日常", *contents, **kwargs):
    """生成报告：标题（必选）、类型（默认日常）、内容（任意数量）、额外配置（任意关键字）"""
    print(f"=== {title}（{type}）===")
    # 打印内容
    for idx, content in enumerate(contents, 1):
        print(f"{idx}. {content}")
    # 打印额外配置
    if kwargs:
        print("\n额外配置：")
        for k, v in kwargs.items():
            print(f"- {k}：{v}")

# 调用函数
create_report(
    "周工作总结",  # 位置参数（title）
    "周报",       # 覆盖默认参数（type）
    "完成需求开发", "修复3个bug", "参加技术分享",  # *args（contents）
    author="张三", date="2026-03-15", department="研发部"  # **kwargs
)
```
