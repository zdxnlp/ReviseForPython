# 字符串

1. 空字符串

```python
empty_str = ''
print(len(empty_str))  # 输出：0（长度为0）
```

2. 转义字符

```python
# 常用转义字符
print('换行符：hello\nworld')  # 换行输出 hello 和 world
print('制表符：name\tage')     # 输出 name    age（制表符分隔）
print('转义单引号：I\'m OK')  # 输出 I'm OK
print('转义反斜杠：C:\\Python')# 输出 C:\Python
```

3. 原始字符串

```python
# 原始字符串：\n 不再是换行，而是普通字符
print(r'原始字符串：C:\Python\numpy')  # 输出：C:\Python\numpy
```

4. 字符串的**不可变性**

```python
s = "hello"
# 尝试修改单个字符（报错）
# s[0] = 'H'  # 报错：TypeError: 'str' object does not support item assignment

# 看似修改，实际创建新字符串（地址变化）
print(f"初始地址：{id(s)}")  # 输出：比如 140682345678928
s = s.replace('h', 'H')     # 替换字符，创建新字符串
print(f"修改后地址：{id(s)}")# 输出：比如 140682345679008（地址变了）
print(s)                    # 输出：Hello
```

5. 索引

```python
s = "Python"
# 正向索引
print(s[0])  # 输出：P（第一个字符）
print(s[3])  # 输出：h（第四个字符）
# 反向索引
print(s[-1]) # 输出：n（最后一个字符）
print(s[-3]) # 输出：h（倒数第三个字符）
```

6. 切片
    - start：起始位置（包含）
    - end：结束为止（不包含）
    - step：步长（默认1，负数表示反方向截取）
    - 左闭右开：[start,end)

```python
s = "HelloWorld"
# 基础切片：截取索引 0-4（不包含4）
print(s[0:4])    # 输出：Hell
# 省略start：从开头到end-1
print(s[:5])     # 输出：Hello
# 省略end：从start到末尾
print(s[5:])     # 输出：World
# 步长为2：每隔1个字符截取
print(s[::2])    # 输出：Hlool
# 反向切片：步长为-1（字符串反转）
print(s[::-1])   # 输出：dlroWolleH
```

7. 拼接与重复
    - 拼接：用+连接字符串（仅字符串间可拼接）
    - 用*重复字符串指定次数

```python
# 拼接
a = "Hello"
b = "World"
print(a + " " + b)  # 输出：Hello World

# 重复
print("-" * 20)     # 输出：--------------------（20个横线，常用作分隔符）
```

## 字符串的常用方法
1. 大小写转换：
    - upper():全部大写
    - lower()：全部小写
    - capitalize()：首字母大写，其余小写
    - title()：每个单词首字母大写

```python
print("hello".upper())
```

```python
print("HELLO".lower())
```

```python
print("python".capitalize())
```

```python
print("python is fun".title())
```

2. 去除空白字符
    - strip():去除首尾空白
    - lstrip():去除左侧空白
    - rstrip():去除右侧空白

```python
print(" hello ".strip())
print(" hello ".lstrip())
print(" hello ".rstrip())
```

3. 查找与替换
    - find(sub)：查找子串位置，找不到返回 -1
    - index(sub)：查找子串位置，找不到报错
    - replace(old, new)：替换子串（默认替换所有）

```python
s="hello"; print(s.find("ll"))
s="hello"; print(s.index("ll"))
s="hello"; print(s.replace("l", "L"))
```

4. 分割与拼接
    - split(sep)：按分隔符分割为列表
    - join(iter)：用字符串拼接可迭代对象（列表 / 元组）

```python
s="a,b,c"; print(s.split(","))
lst=['a','b']; print("-".join(lst))
```

5. 类型判断
    - isdigit()：判断是否全为数字
    - isalpha()：判断是否全为字母
    - isspace()：判断是否全为空白字符
    - startswith(prefix)：判断是否以指定前缀开头
    - endswith(suffix)：判断是否以指定后缀结尾

```python
print("123".isdigit())
print("abc".isalpha())
print(" ".isspace())
print("Python".startswith("Py"))
print("Python".endswith("on"))
```

## 字符串格式化
1. 传统%格式化

```python
name = "张三"
age = 25
# %s 字符串，%d 整数，%f 浮点数
print("姓名：%s，年龄：%d" % (name, age))  # 输出：姓名：张三，年龄：25
print("身高：%.2f 米" % 1.75)            # 输出：身高：1.75 米（保留2位小数）
```

2. str.format()方法

```python
name = "张三"
age = 25
# 位置占位符
print("姓名：{}，年龄：{}".format(name, age))  # 输出：姓名：张三，年龄：25
# 索引占位符
print("姓名：{0}，年龄：{1}".format(name, age))# 输出同上
# 关键字占位符
print("姓名：{n}，年龄：{a}".format(n=name, a=age))# 输出同上
```

3. f-string

```python
name = "张三"
age = 25
# 嵌入变量
print(f"姓名：{name}，年龄：{age}")  # 输出：姓名：张三，年龄：25
# 嵌入表达式
print(f"明年年龄：{age + 1}")       # 输出：明年年龄：26
# 格式化浮点数
height = 1.75
print(f"身高：{height:.2f} 米")    # 输出：身高：1.75 米
```
