- 打擂台的方式，用一个dict存储{当前擂台上面的值:出现的次数}
    1. 如果同样就+1
    2. 如果不一样就-1
    3. 如果擂台上面的为0，新上擂台的代替掉原来守擂台的，次数重置为1

```python
from typing import List


class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        current = nums[0]
        stage = {current:1}
        for i in range(1, len(nums)):
            if stage[current] == 0:
                current = nums[i]
                stage[current] = 1
            else:
                if nums[i] == current:
                    stage[current] += 1
                else:
                    stage[current] -= 1
        return current

if __name__ == '__main__':
    solution = Solution()
    print(solution.majorityElement([2,2,1,1,1,2,2]))
```
