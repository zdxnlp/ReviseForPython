1. 先排序
2. a + b + c = 0，设置三个指针，a每次固定指向一个数，b从前往后遍历，c从后往前遍历。每次判断b+c与-a的大小。
2. \> -a:说明c指向的值大了，故而c向前移动，b不动
3. \< -a:说明b的指向太小了，故而b向后移动，c不动

```python

class Solution:
    def threeSum(self, nums: list[int]) -> list[list[int]]:
        sorted_nums = sorted(nums)
        print(sorted_nums)
        res = []
        a = 0 # 每次固定一个
        b = 1 # 后向遍历
        c = len(nums) - 1
        last_a = -1
        while a < len(nums) - 2: # 前向遍历 如果固定的位置后面仅有1个值了 那么也不存在了 所以 - 2
            current = [] # 存放当前的组合
            if last_a !=-1 and sorted_nums[last_a] == sorted_nums[a]:
                a += 1
                b = a+1
                c = len(nums) - 1
                continue
            while b < c: # 当后向指针在前向指针之前时循环寻找 当b=c时，预示着没找到
                if sorted_nums[a] + sorted_nums[b] + sorted_nums[c] == 0: # 找到了
                    print(f'当前下标：{a}:{sorted_nums[a]},{b}:{sorted_nums[b]},{c}:{sorted_nums[c]}')
                    current += [sorted_nums[a], sorted_nums[b], sorted_nums[c]]
                    res.append(current) # 将当前的组合存进返回值中
                    current = [] # 清空
                    if sorted_nums[b] == sorted_nums[b+1]:
                        while b<c and sorted_nums[b] == sorted_nums[b+1]:
                            b+=1
                    if b<c and sorted_nums[b] != sorted_nums[b+1]:
                        b += 1
                    if sorted_nums[c] == sorted_nums[c-1]:
                        while b<c and sorted_nums[c] == sorted_nums[c-1]:
                            c-=1
                    if b<c and sorted_nums[c] != sorted_nums[c-1]:
                        c -= 1
                    last_a = a
                if sorted_nums[b] + sorted_nums[c] > - sorted_nums[a]:
                    c -= 1
                if sorted_nums[b] + sorted_nums[c] < - sorted_nums[a]:
                    b += 1
            a += 1 # a向后移动
            b = a + 1 # 重置指针
            c = len(nums) - 1 # 重置指针
        return res
if __name__ == '__main__':
    nums = [34,55,79,28,46,33,2,48,31,-3,84,71,52,-3,93,15,21,-43,57,-6,86,56,94,74,83,-14,28,-66,46,-49,62,-11,43,65,77,12,47,61,26,1,13,29,55,-82,76,26,15,-29,36,-29,10,-70,69,17,49]
    Solution().threeSum(nums)
```
