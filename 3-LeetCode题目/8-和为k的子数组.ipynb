```python
from typing import List


class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        position = 0
        sum = 0
        while position < len(nums):
            if nums[position] == k:
                sum += 1
                pointer = position + 1
                while pointer < len(nums):
                    if nums[pointer] == 0:
                        sum += 1
                        pointer += 1
                    else:
                        break
            if position == len(nums) - 1 and nums[position] < k:
                break
            if nums[position] != k:
                print(1)
                pointer = position + 1
                current_k = k - nums[position]
                while pointer < len(nums):
                    if nums[pointer] == current_k:
                        sum += 1
                        current_k = 0
                        pointer += 1
                        continue
                    current_k = current_k - nums[pointer]
                    pointer += 1
            position += 1
        return sum
if __name__ == '__main__':
    solution = Solution()
    print(solution.subarraySum([0,0], 0))
```

```python
from typing import List

class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        """前缀和+哈希表的方式使得时间复杂度为O(n)"""
        count = 0 # 记录一共有多少个连续子串和为k
        prefix = 0 # 当前前缀和
        freq = {0: 1} # 哈希表，记录某个前缀和出现了几次

        for num in nums:
            prefix += num # 修改当前前缀和
            count += freq.get(prefix - k, 0)
            freq[prefix] = freq.get(prefix, 0) + 1

        return count
```
