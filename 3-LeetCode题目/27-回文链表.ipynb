```python
from pyparsing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        if not head:
            return True
        else:
            if head.next is None:
                return True
        pos = head
        len = 0
        while pos:
            len += 1
            pos = pos.next
        stack = []
        for i in range(len // 2):
            stack.append(head.val)
            head = head.next
        if len % 2 != 0:
            head = head.next
        while head:
            if stack.pop() == head.val:
                head = head.next
            else:
                return False
        return True

```
