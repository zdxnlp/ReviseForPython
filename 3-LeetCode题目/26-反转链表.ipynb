```python
from pyparsing import Optional


# 头插法完成链表反转
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head:
            return head
        pos = head
        if head is not None:
            head = head.next
        pos.next = None
        if not head:
            return pos
        while head:
            current = head
            head = head.next
            current.next = pos
            pos = current
        return pos
```
