1. 暴力：从头开始一次向左搜索

```python
class Solution:
    def notIn(self,l:list,s:str):
        length = len(l)
        i = 0
        while i < length:
            if l[i] == s:
                return False
            i += 1
        return True

    def lengthOfLongestSubstring(self, s: str) -> int:
        call = False
        length = str.__len__(s)
        if length == 0:
            return 0
        if length == 1:
            return 1
        maxLenList = []
        maxLen = 0
        i = 0
        j = 1
        lst = []
        lst.append(s[i])
        size = 1
        while j < length:
            if self.notIn(lst,s[j]):
                lst.append(s[j])
                j += 1
                size += 1
                maxLen = max(size, maxLen)
            else:
                if maxLen < size:
                    call = True
                    for k in range(len(lst)):
                        maxLenList.append(lst[i])
                    maxLen = size
                i += 1
                j = i+1
                lst.clear()
                lst.append(s[i])
                size = 1
        return maxLen

if __name__ == '__main__':
    test = Solution()
    print(test.lengthOfLongestSubstring("bb"))
```
