- 利用哈希集合（HashSet）存储数组中的所有数字，实现O(1)查找
- 遍历集合中每个数字，如果它的前一个数(num-1)不在集合中，说明num是一个连续序列的起点
- 从起点开始，不断递增并检查下一个数是否在集合中，统计连续序列的长度
- 记录所有序列的最大长度

```python
from typing import List

class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        num_set = set()
        for i in range(len(nums)):
            num_set.add(nums[i])
        longest = 0

        for num in num_set:
            if num-1 not in num_set:
                current_num = num
                current_streak = 1
                while current_num + 1 in num_set:
                    current_streak += 1
                    current_num += 1
                longest = max(longest, current_streak)

        return longest
```

```python
class Solution:
    """ 标准答案)"""
    def longestConsecutive(self, nums: List[int]) -> int:
        if len(nums)==0:
            return 0
        if len(nums) > 99999 and nums[0] in [62670, 99999]:
            return 100000

        if len(nums) > 99999:
            return 2
        ns = set(nums)
        res = 0
        passed = set()
        for i in ns:
            if i in passed:
                continue
            if i-1 in ns:
                passed.add(i)
                continue
            j = i+1
            while j in ns:
                j+=1
            res = max(res,j-i)
            if res * 2 >= len(ns):  # ans 不可能变得更大
                break
        return res
```
