```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        # 用哈希表存储原来的数组
        table = {}
        res = []
        position = 0
        for i in nums:
            table[i] = position
            position += 1
        new = sorted(nums)
        front = 0
        tail = len(new) - 1
        while front < tail:
            if new[front]+new[tail] == target:
                if new[front] == new[tail]:
                    for i in range(len(nums)):
                        if nums[i] == new[front]:
                            res.append(i)
                    return res
                res.append(table.get(new[front]))
                res.append(table.get(new[tail]))
                break
            if new[front]+new[tail] < target:
                front += 1
            if new[front]+new[tail] > target:
                tail -= 1
        return res

if __name__ == '__main__':
    sol = Solution()
    print(sol.twoSum([1,1], 2))
```
