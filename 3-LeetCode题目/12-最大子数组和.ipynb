````python
from typing import List


class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        s = nums[0]
        p = 1
        m = nums[0]
        while p<len(nums):
            if nums[p] + s < nums[p]:
                s = nums[p]
                m = max(m,s)
            else:
                s += nums[p]
                m = max(m,s)
            p += 1
        return m

if __name__ == '__main__':
    sol = Solution()
    print(sol.maxSubArray([1]))
````
