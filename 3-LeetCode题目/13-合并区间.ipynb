- 按第一个数字排序，构建一个行号表，存储第一个数字从小到大的行号
- 根据行号进行遍历，一次取出两个区间，进行合并。
- 时间复杂度取决于排序算法的时间复杂度。空间复杂度为O(n)

```python
from typing import List

class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        # 按照每行第一个元素进行排序
        sort = sorted(intervals, key=lambda x: x[0])
        res = []
        current = sort[0]
        for position in range(1, len(sort)):
            if sort[position][0] > current[1]:
                res.append(current)
                current = sort[position]
            else:
                current[1] = max(current[1], sort[position][1])
        res.append(current)
        return res


if __name__ == '__main__':
    s = Solution()
    print(s.merge([[1,3],[8,9],[2,6],[4,5]]))
```
