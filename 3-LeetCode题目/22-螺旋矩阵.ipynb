```python
from typing import List

class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
        sum = len(matrix) * len(matrix[0])
        res = []
        r = 1
        up = 0
        down = len(matrix) - 1
        left = 0
        right = len(matrix[0]) - 1
        count = 0
        while True:
            if r == 1: # 右走
                for i in range(left,right+1,1): # 从左到右步长为1
                    res.append(matrix[up][i])
                    count += 1
                if count == sum:
                    break
                up += 1
                r = 2
                continue
            if r == 2: # 向下
                for i in range(up,down+1,1): # 从上到下步长为1
                    res.append(matrix[i][right])
                    count += 1
                if count == sum:
                    break
                right -= 1
                r = 3
                continue
            if r == 3: # 向左
                for i in range(right,left-1,-1): #从右到左步长为1
                    res.append(matrix[down][i])
                    count += 1
                if count == sum:
                    break
                down -= 1
                r = 4
                continue
            if r ==4: # 向上
                for i in range(down,up-1,-1):
                    res.append(matrix[i][left])
                    count += 1
                if count == sum:
                    break
                left += 1
                r = 1
                continue
        return res

if __name__ == '__main__':
    s = Solution()
    print(s.spiralOrder([[1,2,3],[4,5,6],[7,8,9]]))
```
