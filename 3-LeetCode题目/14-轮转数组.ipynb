```python
from typing import List


class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        if k > len(nums):
            k = k%len(nums)
        def reverse(start, end):
            while start < end:
                tmp = nums[start]
                nums[start] = nums[end]
                nums[end] = tmp
                start += 1
                end -= 1
        s1 = 0
        e1 = len(nums) - k - 1
        s2 = len(nums) - k
        e2 = len(nums) - 1
        reverse(s1, e1)
        reverse(s2, e2)
        reverse(s1, e2)
```
