```python
from typing import List

class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        fast = 0
        slow = 0
        while fast < len(nums) and slow < len(nums):
            if fast == len(nums)-1 and nums[fast] == 0: break
            if nums[slow]==0:
                while fast < len(nums):
                    if nums[fast] != 0 and fast > slow:
                        nums[slow], nums[fast] = nums[fast], nums[slow]
                        break
                    if fast == len(nums)-1 and nums[fast] == 0: break
                    fast += 1
            slow += 1

if __name__ == '__main__':
    solution = Solution()
    lst = [1,0]
    solution.moveZeroes(lst)
    print(lst)
```
