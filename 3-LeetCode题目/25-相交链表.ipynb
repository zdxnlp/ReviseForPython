- 先将较长的那个链表的指针往后面移动，使其指向的位置后面的元素个数与较短的一致。然后再逐一进行比较

```python
from typing import Optional


# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        pos = headA
        lenA = 0
        while True:
            if pos:
                lenA += 1
                pos = pos.next
            else:
                break
        pos = headB
        lenB = 0
        while True:
            if pos:
                lenB += 1
                pos = pos.next
            else:
                break
        while lenA > lenB:
            headA = headA.next
            lenA -= 1
        while lenA < lenB:
            headB = headB.next
            lenB -= 1
        while headA and headB and headA != headB:
            headA = headA.next
            headB = headB.next
        if headA:
            return headA
        else:
            return None
```
