```python
from typing import List

class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        res = []
        left = [1,nums[0]] # 左累计乘积已知
        right = [1,nums[-1]] # 右累计乘积已知
        for i in range(2,len(nums)):
            current = left[i-1]
            left.append(nums[i-1]*current)
        current = nums[-1]
        i = len(nums)-3
        while i >= 0:
            current = right[-1]
            right.append(current*nums[i+1])
            i -= 1
        i = 0
        j = len(nums)-1
        while j >= 0:
            res.append(left[i]*right[j])
            j -= 1
            i += 1
        return res
if __name__ == '__main__':
    s = Solution()
    print(s.productExceptSelf([1,2,3,4]))
```
