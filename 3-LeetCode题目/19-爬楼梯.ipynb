# 动态规划
- 到达$n$层有两种方式：
    1. 从$n-1$走$1$步到达
    2. 从$n-2$走$2$步到达
- 所以$f(n)=f(n-1)+f(n-2)$
- 边界条件
    1. $f(1)=1$
    2. $f(2)=2$

```python
class Solution:
    def climbStairs(self, n: int) -> int:
        if n == 1:
            return 1
        if n == 2:
            return 2
        n_1, n_2 = 2, 1
        for i in range(3, n + 1):
            n = n_1 + n_2
            n_2 = n_1
            n_1 = n
        return n

if __name__ == '__main__':
    solution = Solution()
    print(solution.climbStairs(44))
```
